```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#importing data

cred = pd.read_excel("/Users/abhisaurav/Desktop/Cardiff MSc/CreditRiskScoring/GermanCreditData.xlsx", sheet_name="Sheet1")
cred.info()
```

```python
#cred1=cred.astype(str)
cat_cols = cred[['Checking','History','Purpose','Savings','Emploed','marital','Coapp','Property','Other','housing','Job','Telephone','Foreign']]
df=list(cat_cols)
cred[df] = cred[df].astype(str)
cred.info()
```

```python
cred['Good'] = cred['Good'].astype('object')
cred.info()
```

```python
cred.drop(["Bad"],axis=1,inplace=True)
cred["Purpose"].replace(["X"],10, inplace=True)
cred
```

```python
plt.figure(figsize=(18,10))
df = cred[['Amount','Age','Duration']]
df.hist()
```

```python
# Checking outliers using Whiskerplots
df=cred[['Amount','Age','Duration']]
df.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=False, figsize=(14,14), title='Whister plot for input variables')
plt.show()
```

```python
plt.figure(figsize=(16,2))
fig = sns.countplot(y="Good", data=cred,color='mediumseagreen')
plt.show()
```

```python
#Splitting data into duration more and less than 12 month

cred1 = cred[cred["Duration"] <= 12]
cred2 = cred[cred["Duration"] > 12]
cred1
cred2
```

```python
fig, axs = plt.subplots(1, 2,sharex=True, figsize=(10,5))
ax = sns.countplot(x=cred1['Good'],ax=axs[0])
ax = sns.countplot(x=cred2['Good'],ax=axs[1])

axs[0].set_title("<= 12 months")
axs[1].set_title("> 12 months")
```

```python
#Splitting the subsets into training and validation sets.
```

```python
cred1_x = cred1.drop(['Good'],axis = 1)
cred1_y = cred1['Good']

from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(cred1_x, cred1_y, test_size=0.3, random_state=101)
```

```python
cred2_x = cred2.drop(['Good'],axis = 1)
cred2_y = cred2['Good']

from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(cred2_x, cred2_y, test_size=0.3, random_state=101)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import f_classif
from numpy import set_printoptions
from sklearn.feature_selection import mutual_info_classif
```

```python
#Feature Selection for Subset 1
```

```python
X_train1_num = X_train1.select_dtypes('number')
X_train1_cat = X_train1.select_dtypes('object')
X_train1_num
```

```python
fs = SelectKBest(score_func=chi2,k='all')
fit=fs.fit(X_train1_cat,y_train1)
X_train_fs = fit.transform(X_train1_cat)

for i in range(len(fs.scores_)):
	print('Feature %d: %f' % (i, fs.scores_[i]))

#select Checking,History,Property
```

```python
fs = SelectKBest(score_func=f_classif,k=2)
fit=fs.fit(X_train1_num,y_train1)
X_train_fs = fit.transform(X_train1_num)

for i in range(len(fs.scores_)):
	print('Feature %d: %f' % (i, fs.scores_[i]))
    
#Select Installp
```
